## FRAUD CASES IN OTHER INDUSTRY

```python
IndustryScope = MyFile[MyFile['fraudulent' ] ==1]
IndustryScope.head()
```

```python
AllFraudDetails = IndustryScope.groupby('industry').sum()
AllFraudDetails.sort_values(by = "fraudulent", ascending = False, inplace = True)                         
display(AllFraudDetails.head(10))
```

**Oil and Energy industry has the highest fraud case according to my result with 29 recorded cases**

```python
IndustryScope.groupby(['industry'])['fraudulent'].sum().plot.bar(figsize=(12,6))
plt.ylabel('Number of Fraudulent Cases')
plt.xlabel('Industries')
plt.title('Fraudulent Industries')
```

**It wasn't the Information Technology and Services Industry after all. Although it was the most ocurring Industry in the data,it was with only 5 fraudulent cases, Leaving the Oil and Energy Sector to be Top Industry with the highest Fraud Cases**
